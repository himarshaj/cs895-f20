# Analysis of board posting frequency

```python
board_stats = {}

heading_crosswalk = {
    "L": "Location",
    "ボード": "Board ID",
    "ボードのタイトル": "Board Title",
    "１時間以内の投稿。": "Posts Per Hour",
    "すべての投稿": "All Posts",
    "ユニークIP": "Unique IPs",
    "作成済み": "Created",
    "B": "Location",
    "Board": "Board ID",
    "Title": "Board Title",
    "PPH": "Posts Per Hour",
    "Total posts": "All Posts",
    "Active users": "Active Users",
    "Tags": "Tags",
    "Active ISPs": "Active ISPs"
}

for filename in os.listdir("../boards-gendata-with-japanese"):
    
    record_datetime = datetime.strptime( filename.replace('.html.json', ''), "%Y-%m-%dT%H:%M:%SZ" )
    
    
    with open("../boards-gendata-with-japanese/" + filename) as f:
        jdata = json.load(f)
    
    
    for record in jdata:
        
        if 'Name' in record:
            # these are not captures of the board page, but something else from 2017
            # was 8chan broken?
            break

#         board_stats.setdefault( record_datetime, {} )
    
        temp_record = {}
    
        for key in record:

            if key in heading_crosswalk:
                fieldname = heading_crosswalk[key]
            else:
                print("unknown heading {} in {}".format(key, filename))
                sys.exit(255)
            
            value = record[key]
            
            if fieldname == 'Board ID':
                value = record[key].replace(' ', '')
                value = value.replace('\uf0b1', '') # this character doesn't print and appears to be trash
            
            if fieldname == 'All Posts':
                
                if value.strip() == '':
                    value = 0
                    
                value = float(value)
                    
            if fieldname == 'Posts Per Hour':

                if value.strip() == '':
                    value = 0

                value = float(value)
                
            if fieldname == 'Location':
                continue
            
            temp_record[fieldname] = value
                
        board_stats.setdefault(temp_record['Board ID'], {})
        
        board_stats[temp_record['Board ID']].setdefault(record_datetime, {})
        
        board_stats[temp_record['Board ID']][record_datetime]['Posts Per Hour'] = \
            temp_record['Posts Per Hour']
        
        board_stats[temp_record['Board ID']][record_datetime]['All Posts'] = \
            temp_record['All Posts']
        
        board_stats[temp_record['Board ID']][record_datetime]['Board Title'] = \
            temp_record['Board Title']
```

```python
def all_posts_graph(board_stats, boardid, wikipedia_events):
    
    post_dates = []
    post_totals = []

    plt.figure(figsize=(20, 10))

    for post_date in sorted(list(board_stats[boardid].keys())):

        post_dates.append(post_date)
        post_totals.append(
            board_stats[boardid][post_date]['All Posts']
        )


    plt.plot(post_dates, post_totals)
    plt.vlines(
        [
            wikipedia_events
        ],
        1500,
        max(post_totals),
        colors="red"
    )
    plt.xlabel("Year", fontsize=18)
    plt.ylabel("Total Posts", fontsize=18)
    plt.title("8chan's {} board total compared with different events".format(boardid), fontsize=24)

    plt.show()
    
    return post_dates, post_totals
```

```python
def pph_graph_per_board(board_stats, boardid, wikipedia_events):

    post_dates = []
    pph_per_date = []
    
    for post_date in sorted(list(board_stats[boardid].keys())):

        post_dates.append(post_date)
        pph_per_date.append(
            board_stats[boardid][post_date]['Posts Per Hour']
        )
    
    plt.figure(figsize=(20, 10))
    plt.plot(post_dates, pph_per_date)
    plt.vlines(
        [
            wikipedia_events
        ],
        0,
        max(pph_per_date),
        colors="red"
    )
    plt.ylabel("Number of people per hour", fontsize=18)
    plt.xlabel("Year", fontsize=18)
    plt.title("8chan's {} board people per hour compared with different events".format(boardid), fontsize=24)
    plt.show()
    
    return post_dates, pph_per_date
```

## Analysis of ranking of boards each day

```python
board_ranks = {}

heading_crosswalk = {
    "L": "Location",
    "ボード": "Board ID",
    "ボードのタイトル": "Board Title",
    "１時間以内の投稿。": "Posts Per Hour",
    "すべての投稿": "All Posts",
    "ユニークIP": "Unique IPs",
    "作成済み": "Created",
    "B": "Location",
    "Board": "Board ID",
    "Title": "Board Title",
    "PPH": "Posts Per Hour",
    "Total posts": "All Posts",
    "Active users": "Active Users",
    "Tags": "Tags",
    "Active ISPs": "Active ISPs"
}

for filename in os.listdir("../boards-gendata-with-japanese"):
    
    record_datetime = datetime.strptime( filename.replace('.html.json', ''), "%Y-%m-%dT%H:%M:%SZ" )
    
    allboards = []
    
    with open("../boards-gendata-with-japanese/" + filename) as f:
        jdata = json.load(f)
    
    for record in jdata:
        
        temp_record = {}
        
        if 'Name' in record:
            # these are not captures of the board page, but something else from 2017
            # was 8chan broken?
            break

#         board_stats.setdefault( record_datetime, {} )
            
        for key in record:

            if key in heading_crosswalk:
                fieldname = heading_crosswalk[key]
            else:
                print("unknown heading {} in {}".format(key, filename))
                sys.exit(255)
            
            value = record[key]
            
            if fieldname == 'Board ID':
                value = record[key].replace(' ', '')
                value = value.replace('\uf0b1', '') # this character doesn't print and appears to be trash
            
            if fieldname == 'All Posts':
                
                if value.strip() == '':
                    value = 0
                    
                value = float(value)
                    
            if fieldname == 'Posts Per Hour':

                if value.strip() == '':
                    value = 0

                value = float(value)
                
            if fieldname == 'Location':
                continue
            
            temp_record[fieldname] = value

        boardid = temp_record['Board ID']
        pph = temp_record['Posts Per Hour']
      
        allboards.append( (pph, boardid) )
    
    rankcounter = 0
    
    for entry in sorted(allboards, reverse=True):
        
        boardid = entry[1]
        
        board_ranks.setdefault(boardid, {})
        board_ranks[boardid].setdefault(record_datetime, {})
        board_ranks[boardid][record_datetime]["rank"] = rankcounter
        
        rankcounter += 1
```
